```python
#Conversion de datos no numericos a numéricos
columns_to_convert = ['MRP Old', 'Ajio MRP', 'Amazon MRP', 'Amazon FBA MRP', 'Flipkart MRP', 'Limeroad MRP', 'Myntra MRP', 'Paytm MRP']

for column in columns_to_convert:
    datos[column] = pd.to_numeric(datos[column], errors='coerce')
    datos[column] = datos[column].astype(float)
```

```python
#informacion de cada columna despues de la conversion de datos no numericos a numéricos
datos.info()
```

```python
# completar la conversion de datos no numericos a numéricos
columns_to_convert = ['Weight', 'TP','Final MRP Old','Snapdeal MRP']

for column in columns_to_convert:
    datos[column] = pd.to_numeric(datos[column], errors='coerce')
    datos[column] = datos[column].astype(float)
```

```python
#informacion de cada columna despues de la conversion de datos no numericos a numéricos
datos.info()
```

```python
#conocer los datos de la columna no numerica Category , 
unique_values = datos['Category'].unique()
print(f"Valores únicos en la columna '{'Category'}': {unique_values}")
```

```python
#conversion del valor Nill por valor nulo.
datos.replace('Nill', np.nan, inplace=True)
```

```python
#conocer los datos de la columna no numerica Category , 
unique_values = datos['Category'].unique()
unique_values = datos['Category'].unique()
print(f"Valores únicos en la columna '{'Category'}': {unique_values}")
```

```python
#informacion de cada columna despues de la conversion de datos Nill a nan
datos.info()
```

```python
duplicates = datos.duplicated()

# Filtrar el DataFrame para mostrar solo las filas duplicadas
duplicated_rows = datos[duplicates]

print("Filas duplicadas:")
print(duplicated_rows)
```

```python
  # Eliminar las filas con nulos del DataFrame 'datos'
datos.dropna(inplace=True)
```

```python
datos.info()
```

```python
# Conocer los valores unicos de la columna sku
unique_values = datos['Sku'].unique()
print(f"Valores únicos en la columna '{'Sku'}': {unique_values}")
```

```python
# Creacion de columna Size con la parte final de la columna SKU
datos['Size'] = datos['Sku'].str.rsplit('_', n=1).str[1]
```

```python
datos.info()
```

```python
# Conocer los valores unicos de la columna Size
unique_values = datos['Size'].unique()
print(f"Valores únicos en la columna '{'Size'}': {unique_values}")
```

```python
datos.count()
```

```python
# Obtener el índice de la columna "Style Id"
style_id_index = datos.columns.get_loc('Style Id')

# Verificar si la columna 'Size' está en la lista de columnas
if 'Size' in datos.columns:
    # Obtener las columnas en el orden deseado
    new_column_order = datos.columns.tolist()
    new_column_order.remove('Size')
    new_column_order.insert(style_id_index + 1, 'Size')
    
# Reordenar las columnas del DataFrame
    datos = datos[new_column_order]
```

```python
# Nueva data posterior a la limpieza
datos
```

```python
# Convertir el DataFrame a un archivo CSV
nombre_archivo = 'Mayo-2022 2v.csv'
datos.to_csv(nombre_archivo, index=False)
```
