```python
import numpy as np
import pandas as pd
import seaborn as sns
import glob 
import warnings
import matplotlib.pyplot as plt
import matplotlib.style as style
from datetime import datetime
ei=pd.read_csv("C:/Users/felip/Desktop/Master_BI_DM/Data Set/07 Expense IIGF.csv")
ei
```

```python
## Cambiar nombres a columnas
ei.columns = ['index','Received Date', 'Received Amount', 'Expance particular', 'Expance Amount']
ei
```

```python
## Eliminamos las filas no relevantes
ei = ei.drop(ei[ei['index']>=14].index)
ei
```

```python
#eliminamos la fila 0
ei.drop(0,inplace = True)
ei
```

```python
# Eliminamos la columba "index"
ei.drop("index", axis = 1,inplace = True)
ei
```

```python
## Dar formato Fecha a columna DATE
ei['Received Date'] = pd.to_datetime(ei['Received Date'])
print(ei["Received Date"])
```

```python
#reemplazar con cero las fecha NaT
ei['Received Date'] = ei['Received Date'].fillna(0)
ei
```

```python
#remplazar con cero los valores NaN
ei['Received Amount'] = ei['Received Amount'].fillna(0)
ei
```

```python
ei.info()
```

```python
#dar formato a columnas con datos numericos
ei['Received Amount'] = ei['Received Amount'].astype('int64')
ei['Expance Amount'] = ei['Expance Amount'].astype('int64')
ei.info()
```

```python
# Calculamos la suma para ciertas las columnas con valores.
ei[['Received Amount', 'Expance Amount']].sum()
```
