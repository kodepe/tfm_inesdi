```python
## este df indica el stock de cada SKU. Consotruiremos un nuevo archivo.csv con el stock.
## ver nombres de columnas para cambiarla
isr2.columns
```

```python
##cambiar nombres
isr2.columns = ['index', 'CUSTOMER', 'DATE', 'Months', 'Style', 'SKU','PCS','RATE','GROSS AMT','Stock']
isr2
```

```python
## elimina primer registro que tiene el nombre de la columna
isr2.drop([0], axis=0, inplace=True)
isr2
```

```python
##reviso si hay datos nulos en esta parte del dataset
round(isr2.isnull().sum() / isr2.shape[0] * 100.00,2)
```

```python
## Elimino los Nulos
isr2=isr2.dropna()
isr2.info()
```

```python
# FunciÃ³n para extraer la talla de la columna SKU
def extraer_talla(SKU):
    partes = SKU.split('-')
    if len(partes) >= 2:
        return partes[-1]
    else:
        return None

# Aplicar la funciÃ³n a la columna SKU y crear una nueva columna 'talla'
isr2['Size'] = isr2['SKU'].apply(extraer_talla)
```

```python
# revisar los datos de cada columna
for col in isr2.columns:
    print("Column:",col, "\n\n", isr2[col].value_counts(),"\n")
```
